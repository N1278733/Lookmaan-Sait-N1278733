# Energy level
In the dataset, the average of the energy column of the dataset can be calculated by iterating the data using the looping statement. By using the for loop statement of the Python, we can calculate the sum popularity and using this sum we can calculate the average. The result can be used to identified what is energy of a person while listening the spotify songs

```python
spotify_data = open('Spotify-playlist-origins.csv','r')
spotify_rows=csv.DictReader(spotify_data)
count = 0
sum = 0
for row in spotify_rows:
    sum = sum + float(row['Energy'])
    count = count + 1
    
average_energy = sum/count
print('The average energy is {}'.format(average_energy))
```

# The count of songs released after 2020
The total number of songs released after the year 2020 can be calculated by using the 'Release Date' column. The conditional statement is used to check the relase date, whether it is after the year of 2020 or not. Each rows are iterated and checked with the if statement.

```python
from datetime import date
from datetime import datetime
spotify_data = open('Spotify-playlist-origins.csv','r')
spotify_rows=csv.DictReader(spotify_data)
date_format = '%d-%m-%Y'
req_date = datetime.strptime('01-01-2020', date_format)
count = 0
for row in spotify_rows:
    release_date = datetime.strptime(row['Release Date'], date_format)
    if release_date > req_date: 
        count = count + 1
print('The total number of songs released after the year of 2020 is {}'.format(count))
```

# GUI Development 

The GUI can be developed using the tkinter module in Python programing language. We can give an interaction window for choosing the type of visualization a user want to view. Once it is selected, the corresponding visualization will be displayed on this notebook. This can be achieved by giving visualization plot through the radio button. To achieve this RadioButton class from tkinter module can be used. In which, we can give the parameters for variable, value, command. Finally the instance will be packed together for all the radio buttons then it will be displayed on the screen using mainloop() method.

The selection will be tested in viewSelectedVisual() method and using the conditional statement, the corresponding visualization can be displayed on the screen.

```python
from tkinter import *

ws = Tk()
ws.title('Data Visualization')
ws.geometry('300x220')

def viewSelectedVisual():
    choice  = selection.get()
    ws.destroy()
    sns.color_palette("tab10")
    if choice == 1:
           sns.lineplot(x=spotify_dataframe['Loudness'],y=spotify_dataframe['Liveness']).set(title='Song Quality')
    elif choice == 2:
       sns.scatterplot(x=spotify_dataframe['Loudness'],y=spotify_dataframe['Liveness']).set(title='Song Quality')    
    elif choice == 3:
       sns.boxplot(x=spotify_dataframe['Speechiness']).set(title='Speechiness')
    elif choice == 4:
        sns.violinplot(x=spotify_dataframe['Tempo']).set(title='Flow of Tempo')    
    elif choice == 5:
        sns.swarmplot(x=spotify_dataframe['Tempo']).set(title='Flow of Tempo')
    elif choice == 6:
        sns.barplot(x=spotify_dataframe['Acousticness'],y=spotify_dataframe['Instrumentalness']).set(title='Instrument Usage')
    elif choice == 7:
        sns.pointplot(x=spotify_dataframe['Acousticness'],y=spotify_dataframe['Instrumentalness']).set(title='Instrument Usage')    
    
selection = IntVar()
Label(ws,text='Select a visualization method').pack()
Radiobutton(ws, text="Line Plot", variable=selection, value=1, command=viewSelectedVisual).pack()
Radiobutton(ws, text="Scatter Plot", variable=selection, value=2, command=viewSelectedVisual).pack()
Radiobutton(ws, text="Box Plot", variable=selection, value=3, command=viewSelectedVisual).pack()
Radiobutton(ws, text="Violin Plot", variable=selection, value=4, command=viewSelectedVisual).pack()
Radiobutton(ws, text="Swarm Plot", variable=selection, value=5, command=viewSelectedVisual).pack()
Radiobutton(ws, text="Bar Plot", variable=selection, value=6, command=viewSelectedVisual).pack()
Radiobutton(ws, text="Point Plot", variable=selection, value=7, command=viewSelectedVisual).pack()

ws.mainloop()
```
